```python
plt.figure(figsize=(20,6))
plt.hist(kaggle['Duration'].astype(float));
```

```python
plt.figure(figsize=(10,6))
plt.bar( kaggle.gender.unique(),kaggle['gender'].value_counts(),color = '#09c')
plt.show()
```

```python
age_count = kaggle['age'].value_counts()
plt.figure(figsize=(10,6))
ax = plt.axes()
ax.set_facecolor("white")
plt.bar( sorted(kaggle.age.unique()),age_count,color = '#09c')
plt.show()
```

```python
plt.pie(kaggle.groupby('Duration').count()['gender'][:2],labels=kaggle['gender'].unique()[:2],explode=[0,.05,],autopct='%1.1f%%');
```

```python
kaggle['programming_language_Swift'] = kaggle['programming_language_Swift'].replace(np.nan,'not_Swift')
count_Java = kaggle.groupby(['gender','programming_language_Swift']).count()['Duration']
fig_dims = (10, 5)
fig, ax = plt.subplots(figsize=fig_dims)
BAR = sb.barplot(x= 'gender', y='Duration',hue = 'programming_language_Swift',ax = ax,data=kaggle)
BAR.set(xlabel='Gender', ylabel='Numbers per gender')
```
